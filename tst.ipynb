```python
import pandas as pd
import numpy as np
from pprint import pprint
import matplotlib.pyplot as plt
```

```python
#read in csv file
data = pd.read_csv("SpotifyTopSongsByCountry.csv")
```

```python
#view data csv
data.head()
```

```python
#create a dataframe with only Artists column
artists_df = data[['Artists']]
artists_df
```

```python
#clean data to only include one artist name per column (delimiter is a comma)
cleaned_df = artists_df.Artists.str.split(",", expand=True)
cleaned_df.head()
```

```python
#add row[0], first artist name listed, to the orginial dataframe
data['Artist_clean'] = cleaned_df[0]
```

```python
#check to see if first artist column was added
data.head()
```

```python
#get column names for dataframe
for col in data.columns:
    print(col)
```

```python
#reorder columns in dataframe so Artists and First Artist are next to each other
data = data[['Country', 'Continent', 'Rank', 'Title', 'Artists', 'Artist_clean', 'Album', 'Explicit', 'Duration']]
data
```

```python
data.to_csv("tst_cleaned_top_song.csv")
```

```python
#read in csv file
artists = pd.read_csv("data_spot._cleanartistcsv.csv")
```

```python
artists.head()
```

```python
complete_data = pd.merge(data, artists, on="Artist_clean", how="inner")
```

```python
#read in csv file
complete_data.to_csv("complete_data.csv")

complete_data
```

```python
artist_mean = complete_data.groupby(['Artist_clean']).mean()
artist_mean
```

```python
#list artists and number of unique countries they had top songs in
artist_country = pd.DataFrame({"Artist_clean": complete_data['Artist_clean'],
                                "Country": complete_data['Country']
})

artist_country = artist_country.groupby(['Artist_clean']).nunique()[('Country')].sort_values(ascending=False)
artist_country
# artist_country.dtypes
```

```python
artist_country.plot(kind="bar", figsize=(40,5))
plt.ylabel("Country")
plt.xlabel("Artist")
plt.title("Country per Artist")
```
