```python
import pandas as pd
import numpy as np
from pprint import pprint
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv('data_spot._cleanartistcsv.csv')
df_top_songs = pd.read_csv('SpotifyTopSongsByCountry.csv')
```

```python
df_drop = df.drop(columns = ['Artists','Id','Mode','Key'])
```

```python
df_drop.columns
```

```python
df_arrange = df_drop[['Artist_clean','Release Date','Year','Acousticness','Danceability',
       'Duration Ms', 'Energy', 'Instrumentalness', 'Liveness', 'Loudness','Popularity', 'Speechiness', 'Tempo', 'Valence','Explicit','Number of Records']]
```

```python
df_arrange.columns
```

```python
df_rename = df_arrange.rename(columns={"Artist_clean": "Artist",
                                      "Year": "Release_Year", "Number of Records": "Records"})
```

```python
df_spotify = df_rename
```

```python
df_spotify.head()
```

```python
df_spotify_gb = df_spotify.groupby(['Artist']).mean()
```

```python
df_spotify_gb.head(20)
```

```python
df_top_songs.columns
```

```python
df_songs_col = df_top_songs[['Artists','Album','Continent','Country','Rank']]

df_songs_col.head()
```

```python
df_songs_col.head()
```

```python
artists_df = df_top_songs[['Artists']]
artists_df
```

```python
artists_songs_df = artists_df.Artists.str.split(",", expand=True)
artists_songs_df.head()
```

```python
df_top_songs['Artist_clean'] = artists_songs_df[0]

df_top_songs.columns
```

```python
clean_top_songs = df_top_songs[['Artist_clean','Album','Continent','Country','Rank',]]
```

```python
comp_data = pd.merge(df_spotify_gb, clean_top_songs, left_on = 'Artist', right_on = 'Artist_clean', how = 'inner')
```

```python
comp_data.head()

comp_drop_glo = comp_data.loc[comp_data['Country']!='Global']

comp_drop_glo['Continent'].nunique()
```

```python
comp_drop_glo.to_csv("complete_data_EM_3.csv")
```
